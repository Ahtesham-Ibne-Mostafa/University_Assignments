```python
class Node:
    def __init__(self, e, n, p):
        self.element = e
        self.next = n
        self.prev = p
```

```python
class DoublyList:
    
    def __init__(self, a):
        if type(a) == list:
            self.head = None
            self.tail = None
            x = len(a) - 1
            for i in range(0, len(a)):
                n = Node(a[x], self.head, None)
                x -= 1
                if n.next == None:
                    tail = n
                    self.head = n
                    self.tail = n
                else:
                    self.head = n
                    tail.prev = n
                    tail = n
                    
    def countNode(self):
        h = self.head
        count = 0
        while h != None:
            count += 1
            h = h.next
        return count
    
    def forwardprint(self):
        h = self.head
        string = ""
        while h != None:
            if h.next != None:
                string += str(h.element) + ", "
            else:
                string += str(h.element) + "."
            h = h.next
        print(string)
    
    def backwardprint(self):
        t = self.tail
        string = ""
        while t != None:
            if t.prev != None:
                string += str(t.element) + ", "
            else:
                string += str(t.element) + "."
            t = t.prev
        print(string)
    
    def nodeAt(self, idx):
        h = self.head
        if idx >= 0 and idx < self.countNode():
            for i in range(0, idx):
                h = h.next
            return h
        else:
            return None
    
    def indexOf(self, elmnt):
        h = self.head
        idx = 0
        while h != None:
            if h.element == elmnt:
                return idx
            h = h.next
            idx += 1
        return -1
    
    def insert(self, elmnt, idx):
        h = self.head
        if idx == 0:
            p = None
            nn = Node(elmnt, h, p)
            n1 = self.nodeAt(0)
            self.head = nn
            n1.prev = nn
        elif idx > 0 and idx < self.countNode():
            n1 = self.nodeAt(idx-1)
            n2 = self.nodeAt(idx)
            nn = Node(elmnt, n1.next, n1)
            n1.next = nn
            n2.prev = nn
        elif idx == self.countNode():
            n1 = self.nodeAt(idx-1)
            nn = Node(elmnt, None, None)
            n1.next = nn
            nn.prev = n1
            self.tail = nn
        
    def remove(self, idx):
        h = self.head
        if idx == 0:
            elmnt = h.element
            self.head = h.next
            self.head.prev = None
            h.next = None
            return str(elmnt)
        elif idx == self.countNode() - 1:
            nidx = self.nodeAt(idx)
            n1 = self.nodeAt(idx-1)
            elmnt = nidx.element
            n1.next = None
            nidx.prev = None
            self.tail = n1
            return str(elmnt)
        elif idx > 0 and idx < self.countNode():
            nidx = self.nodeAt(idx)
            n1 = self.nodeAt(idx-1)
            n2 = self.nodeAt(idx+1)
            elmnt = nidx.element
            n1.next = nidx.next
            n2.prev = nidx.prev
            nidx.next = None
            nidx.prev = None
            return str(elmnt)
        else:
            return None
    
        
print("///  Test 01  ///")
a1 = [10, 20, 30, 40]
h1 = DoublyList(a1)
h1.forwardprint()
h1.backwardprint()
print(h1.countNode())

print("///  Test 02  ///")
myNode = h1.nodeAt(2)
print(myNode.element)

print("///  Test 03  ///")
index = h1.indexOf(40)
print(index)

print("///  Test 04  ///")
a2 = [10, 20, 30, 40]
h2 = DoublyList(a2)
h2.forwardprint() 
h2.insert(85,0)
h2.forwardprint()
h2.backwardprint()
print()
h2.insert(95,3)
h2.forwardprint()
h2.backwardprint()
print()
h2.insert(75,6)
h2.forwardprint()
h2.backwardprint()

print("///  Test 05  ///")
a3 = [10, 20, 30, 40, 50, 60, 70]
h3 = DoublyList(a3)
h3.forwardprint()
print("Removed element: "+ h3.remove(0))
h3.forwardprint() 
h3.backwardprint()
print("Removed element: "+ h3.remove(3))
h3.forwardprint()
h3.backwardprint()
print("Removed element: "+ h3.remove(4))
h3.forwardprint()
h3.backwardprint()
```
