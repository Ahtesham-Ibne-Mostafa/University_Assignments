```python
#task 1
# Write your code here
class PlayerEarning:
    def __init__(self,name):
        self.name = name
        self.earning = 0
        self.earning_with_bonus = 0
        self.bonus = 0
    def calculateTotal(self,earning,goal=None):
        self.earning = earning
        if goal==None:
            pass
        elif goal>30:
            self.bonus = (5/100)*earning+10000
        else:
            self.bonus = (5/100)*earning
        self.earning_with_bonus = self.earning + self.bonus    
    def printDetails(self):
        print("Player Name:",self.name)
        print("Player Season Earning without bonus:",self.earning)
        print("Bonus:",int(self.bonus))
        print("Player Season Earning After Bonus:",int(self.earning_with_bonus))
print("**********************")
player1 = PlayerEarning('Buffon')
player1.calculateTotal(250000)
player1.printDetails()

print("\n**********************")
player2 = PlayerEarning('Dybala')
player2.calculateTotal(250000, 31)
player2.printDetails()

print("\n**********************")
player3 = PlayerEarning('Cuadrado')
player3.calculateTotal(250000, 20)
player3.printDetails()
```

```python
#Task 15
class Player:
    database={}
    playerNo = 0
    def __init__(self,name,team,jerseyNo):
        self.name = name
        self.team = team
        self.jerseyNo = jerseyNo
    def __str__(self):
        return "Name:{}\nTeam:{}\nJersey No:{}".format(self.name,self.team,self.jerseyNo)

#Write your code here
class FootballPlayer(Player):
    def __init__(self,name,team,jerseyNo,goals,r_d="Not yet retired"):
        super().__init__(name,team,jerseyNo)
        Player.playerNo+=1
        self.goals = goals
        lst1 = name.split()
        first_name = lst1[0]
        second_name = lst1[1]
        self.player_id = str(Player.playerNo)+(first_name[0]+second_name[0]).upper()+str(jerseyNo)
        self.r_d = r_d
        Player.database[self.player_id] = [name,team,jerseyNo,goals,r_d] 
    def createPlayer(name,team,jerseyNo,goals,r_d):
        return FootballPlayer(name,team,jerseyNo,goals,r_d)
    def __str__(self):
        return("Player ID: "+self.player_id+"\n"+super().__str__()+"\nGoals Scored:"+str(self.goals)+"\nRetirement date: "+self.r_d)
    
        
print("Number of players:",Player.playerNo)
print("Player Database:",Player.database)
print("#################################")
p1 = FootballPlayer("Lionel Messi","Barcelona",10,231)
print("------Details of the player------")
print(p1)
print("#################################")
p2 = FootballPlayer("Cristiano Ronaldo","Juventus",7,215)
print("------Details of the player------")
print(p2)
print("#################################")
p3 = FootballPlayer.createPlayer("Miroslav Klose","Lazio",11,71,"11 Aug,2014")
print("------Details of the player------")
print(p3)
print("#################################")
print("Number of players:",Player.playerNo)
print("Player Database:",Player.database)
```

```python
#Task 10
class Student:
    def __init__(self,name,ID):
        self.name = name
        self.ID = ID
    def Details(self):
        return "Name: "+self.name+"\n"+"ID: "+self.ID+"\n"
#Write your code here
class CSEStudent(Student):
    def __init__(self,name,ID,semester):
        super().__init__(name,ID)
        self.semester = semester
        self.course_lst = []
        self.grade_lst = []
        self.gpa = 0
    def Details(self):
        return (super().Details()+"Current semester: "+self.semester)
    def addCourseWithMarks(self,*args):
        
        for index in range(len(args)):
            if index%2==0:
                self.course_lst.append(args[index])
            else:
                if args[index]>=85:
                    grade = 4.0
                elif args[index]>=80 and args[index]<=84:
                    grade = 3.3
                elif args[index]>=70 and args[index]<=79:
                    grade = 3.0
                elif args[index]>=65 and args[index]<=69:
                    grade = 2.3
                elif args[index]>=57 and args[index]<=64:
                    grade = 2.0    
                elif args[index]>=55 and args[index]<=56:
                    grade = 1.3
                elif args[index]>=50 and args[index]<=54:
                    grade = 1.0
                else:
                    grade = 0.0
                self.grade_lst.append(grade)
        self.gpa = (sum(self.grade_lst*3))/(3*len(self.course_lst))
    def showGPA(self):    
        print(self.name,"has taken",len(self.course_lst),"courses.")
        for index in range(len(self.course_lst)):
            print(self.course_lst[index],":",self.grade_lst[index])
        print("GPA for",self.name,"is:",format(self.gpa,".2f"))    
                
Bob = CSEStudent("Bob","20301018","Fall 2020")
Carol = CSEStudent("Carol","16301814","Fall 2020")
Anny = CSEStudent("Anny","18201234","Fall 2020")
print("#########################")
print(Bob.Details())
print("#########################")
print(Carol.Details())
print("#########################")
print(Anny.Details())
print("#########################")
Bob.addCourseWithMarks("CSE111",83.5,"CSE230",73.0,"CSE260",92.5)
Carol.addCourseWithMarks("CSE470",62.5,"CSE422",69.0,"CSE460",76.5,"CSE461",87.0)
Anny.addCourseWithMarks("CSE340",45.5,"CSE321",95.0,"CSE370",91.0)
print("----------------------------")
Bob.showGPA()
print("----------------------------")
Carol.showGPA()
print("----------------------------")
Anny.showGPA()
```
